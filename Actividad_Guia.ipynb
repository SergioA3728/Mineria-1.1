##  Paso 2: Cargar el dataset y revisar estructura

```python
import pandas as pd
df = pd.read_csv('university_student_data.csv')
print('Dataset cargado correctamente')
df.head()
```

## Paso 3: Exploración inicial (EDA)
Verificamos las columnas, tipos de datos y valores nulos.

```python
df.info()
df.describe(include='all')
```

##  Paso 4: Visualizaciones básicas

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,4))
sns.lineplot(x='Year', y="Retention Rate (%)", data=df, marker='o')
plt.title('Tendencia de Retención por Año')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
sns.barplot(x="Year", y="Student Satisfaction (%)", data=df)
plt.title("Satisfacción Promedio por Año")
plt.show()
```

## 🔹 Paso 5: Preparación del Dashboard (Streamlit)
Aquí generamos el archivo **`app.py`** que desplegarás en Streamlit Cloud.

```python
code = open("app.py").read()
with open("app_colab_streamlit.py","w") as f:
    f.write(code)
print("Archivo app_colab_streamlit.py generado correctamente.")
```
